```python
import pandas as pd
from groq import Groq

client = Groq(api_key="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")  


data = pd.read_excel("/content/cleaned_output_code_mixdata.xlsx")


test_data = data.head(20000)  


def get_result_from_llm(prompt):
    try:
       
        chat_completion = client.chat.completions.create(
            messages=[{"role": "user", "content": prompt}],
            model="llama-3.3-70b-versatile",  
            temperature=1,
            max_tokens=4000,
            top_p=1,
            stop=None,
            stream=False,
        )
        return chat_completion.choices[0].message.content
    except Exception as e:
        print(f"Error occurred: {e}")
        time.sleep(5) 
        return get_result_from_llm(prompt)


def translate_text(text, target_language):
    prompt = f"Telugu-English mixed Text: {text}. Translate this code-mixed sentence to {target_language}. Just print the translation alone."
    translated_text = get_result_from_llm(prompt)
    return translated_text

test_data['English_Translation'] = test_data['Text'].apply(lambda x: translate_text(x, 'English'))
test_data['Telugu_Translation'] = test_data['Text'].apply(lambda x: translate_text(x, 'Telugu'))

print(test_data[['Text', 'English_Translation', 'Telugu_Translation']])
```
